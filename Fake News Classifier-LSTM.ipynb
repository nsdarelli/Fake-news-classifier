```python
##Fake News Classifier ##LSTM

import pandas as pd

df = pd.read_csv('/content/train1.csv', error_bad_lines=False, engine='python')
```

```python
#Drop Nan Values
df = df.dropna()
```

```python
# Independent feature
X = df.drop('label',axis=1)
```

```python
#Dependent feature
y=df['label']
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size = 5000
```

```python

messages = X.copy()
```

```python
messages['title'][0]
```

```python
messages
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Data Preprocessing

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

corpus = []

for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z0-9]', ' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[0]
```

```python
#One_hot rep:

onehot_rpr = [one_hot(words, voc_size) for words in corpus]
onehot_rpr
```

```python
corpus[0]
```

```python
onehot_rpr[0]
```

```python
#Embedding Rep: and equal the length of seq with padding

sent_len = 20
embedd = pad_sequences(onehot_rpr, padding='pre', maxlen=sent_len)
print(embedd)
```

```python
from keras.utils.sidecar_evaluator import optimizer
from tensorflow.keras.layers import Dropout
# Creating model

embedd_vec_feat = 40
model = Sequential()
model.add(Embedding(voc_size, embedd_vec_feat, input_length=sent_len))
model.add(Dropout(0.33))
model.add(LSTM(100))
model.add(Dropout(0.33))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
len(embedd), y.shape
```

```python
import numpy as np

X_final = np.array(embedd)
y_final = np.array(y)
```

```python

y_final
```

```python
X_final.shape, y_final.shape
```

```python
#Train Test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
#Model Training

model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=10, batch_size=128)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred> 0.5, 1,0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
